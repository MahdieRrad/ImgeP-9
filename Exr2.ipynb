```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import ImageFont, ImageDraw, Image
```

```python
imge=np.ones((610,1400,3),np.uint8)*60
imge[210:310,210:310]=(200,83,20)
imge[320:420,210:310]=(0,140,200)
imge[210:310,320:420]=(150,200,0)
imge[320:420,320:420]=(255,187,0)
font = ImageFont.truetype('font/seguisb.ttf',160)
image=Image.fromarray(imge)
draw=ImageDraw.Draw(image)
draw.text((480,195),'Microsoft',font=font,fill=(255,255,255))
imge = np.array(image)
plt.imshow(imge)
```
