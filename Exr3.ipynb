```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
imge = cv2.imread('rubc.jpg')
RGB= cv2.cvtColor(imge,cv2.COLOR_BGR2RGB)
plt.imshow(RGB)
```

```python
row ,col = RGB.shape[:2]
for i in range (row):
    for j in range (col):
        if RGB[i,j][0]>50 and RGB[i,j][1]>50 and RGB[i,j][2]<50:
            RGB[i,j]=(0,0,255)
        elif RGB[i,j][0]<50 and RGB[i,j][1]>50 and RGB[i,j][2]>50:
            RGB[i,j] = (255,0,0)
        elif RGB[i,j][0]>50 and RGB[i,j][1]<50 and RGB[i,j][2]>50:
            RGB[i,j]=(0,255,0)
            
            
plt.imshow(RGB)
```
